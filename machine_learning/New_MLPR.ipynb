# Load Data

```python
data = pd.read_csv('/content/drive/MyDrive/ML/shuffled_data.csv')
```

# Traditional Preprocessing

```python
# Additional code -> Changing some features into numerical float
data['RAM'] = data['RAM'].apply(lambda x: float(x.split(' ')[0]))
data['Weight'] = data['Weight'].apply(lambda x: float(x.replace(',','.').split(' ')[0]))
data['Storage'] = data['Storage'].apply(lambda x: float(x.split(' ')[0]))
data['Screen Size'] = data['Screen Size'].apply(lambda x: float(x.replace(',','.').split(' ')[0]))
data['Pixel Density'] = data['Pixel Density'].apply(lambda x: float(x.split(' ')[0]))
data['Battery Capacity'] = data['Battery Capacity'].apply(lambda x: float(x.split(' ')[0]))
data['Price'] = data['Price'].apply(lambda x: float(x.replace('.', '').replace('Rp ', '')))
```

```python
data
```

# Data Preparation

## Separate Categorical and Numerical Data, then combine
- Categorical -> one hot encoding
- Numerical -> Stay the same

```python
# Separate data that contains categorical and regression columns
categorical = data.drop(['Name', 'Price', 'RAM', 'Weight', 'Storage', 'Screen Size', 'Pixel Density', 'Battery Capacity'], axis = 1)
categorical_x = pd.get_dummies(categorical) # one hot encoding categorical data tanpa perlu convert ke numpy
categorical_x_df = pd.DataFrame(categorical_x, columns=categorical_x.columns) # convert ke data frame berdasarkan column categorical_x

regression = data.drop(['Name', 'Brand', 'Chipset', 'CPU', 'GPU', 'Screen Type', 'Resolution', 'Total camera'], axis = 1) # Since bukan data categorical ga perlu diapa-apain
combined_df = pd.concat([categorical_x_df, regression], axis = 1)
```

## Separate Data into X and Y

```python
# RUN THIS
# Seperate data into X and Y
X = combined_df.drop(['Price'], axis = 1).to_numpy()
# X = pd.get_dummies(X).to_numpy() # change into numpy format <-- Yang bapak coding yang one hot encoder
y = combined_df['Price'].to_numpy().reshape(-1,1)
# y = pd.get_dummies(y).to_numpy().reshape(-1,1)
```

## Data Split into Training and Test Data

```python
from sklearn.model_selection import train_test_split
X_train, X_test,y_train, y_test = train_test_split(X,y, test_size=0.2, random_state=0, shuffle=False)
```

## Scaling

### Z-Scaler

```python
from sklearn.preprocessing import StandardScaler
z_scale_x = StandardScaler()
X_train_z = z_scale_x.fit(X_train)
X_train_z = z_scale_x.transform(X_train)
X_test_z = z_scale_x.transform(X_test)

z_scale_y = StandardScaler()
y_train_z = z_scale_y.fit(y_train)
y_train_z = z_scale_y.transform(y_train)
y_test_z = z_scale_y.transform(y_test)
```

### Minmax Scaler

```python
from sklearn.preprocessing import MinMaxScaler
minmax_scale_x = MinMaxScaler()
X_train_minmax = minmax_scale_x.fit(X_train)
X_train_minmax = minmax_scale_x.transform(X_train)
X_test_minmax = minmax_scale_x.transform(X_test)

minmax_scale_y = MinMaxScaler()
y_train_minmax = minmax_scale_y.fit(y_train)
y_train_minmax = minmax_scale_y.transform(y_train)
y_test_minmax = minmax_scale_y.transform(y_test)
```

# Model : MLPR

```python
# Model
from sklearn.neural_network import MLPRegressor

# No Scale
regressor = MLPRegressor(random_state = 0)
regressor.fit(X_train, y_train)
pred = regressor.predict(X_test)

# Standard scaling
regressor_z = MLPRegressor(random_state = 0)
regressor_z.fit(X_train_z, y_train_z)
pred_z = regressor_z.predict(X_test_z)
pred_z_inverse = z_scale_y.inverse_transform(pred_z.reshape(-1,1))

# MinMax scaling
regressor_minmax = MLPRegressor(random_state = 0)
regressor_minmax.fit(X_train_minmax, y_train_minmax)
pred_minmax = regressor_minmax.predict(X_test_minmax)
pred_minmax_inverse = minmax_scale_y.inverse_transform(pred_minmax.reshape(-1,1))
```

# Evaluation

```python
# MAE
from sklearn.metrics import mean_absolute_error as mae
print("MEAN ABSOLUTE ERROR")
# No Scaling
print("MAE                :", mae(y_test, pred))

# Standard Scaling
print("Standard Scaling")
print("MAE (On Scale)     :", mae(y_test_z, pred_z))
print("MAE (Inverse Scale):", mae(z_scale_y.inverse_transform(y_test_z), pred_z_inverse))

# Minmax Scaling
print("Minmax Scaling")
print("MAE (On Scale)     :", mae(y_test_minmax, pred_minmax))
print("MAE (Inverse Scale):", mae(minmax_scale_y.inverse_transform(y_test_minmax), pred_minmax_inverse))

# MSE
from sklearn.metrics import mean_squared_error as mse
print("MEAN SQUARED ERROR")
# No Scaling
print("MSE                :", mse(y_test, pred))

# Standard Scaling
print("Standard Scaling")
print("MSE (On Scale)     :", mse(y_test_z, pred_z))
print("MSE (Inverse Scale):", mse(z_scale_y.inverse_transform(y_test_z), pred_z_inverse))

# Minmax Scaling
print("Minmax Scaling")
print("MSE (On Scale)     :", mse(y_test_minmax, pred_minmax))
print("MSE (Inverse Scale):", mse(minmax_scale_y.inverse_transform(y_test_minmax), pred_minmax_inverse))

# MRSE
print("MEAN SQUARED ROOT ERROR")
# No Scaling
print("MSRE                :", mse(y_test, pred, squared = False))

# Standard Scaling
print("Standard Scaling")
print("MSRE (On Scale)     :", mse(y_test_z, pred_z, squared = False))
print("MSRE (Inverse Scale):", mse(z_scale_y.inverse_transform(y_test_z), pred_z_inverse, squared = False))

# Minmax Scaling
print("Minmax Scaling")
print("MSRE (On Scale)     :", mse(y_test_minmax, pred_minmax, squared = False))
print("MSRE (Inverse Scale):", mse(minmax_scale_y.inverse_transform(y_test_minmax), pred_minmax_inverse, squared = False))

# MAPE
from sklearn.metrics import mean_absolute_percentage_error as mape
print("MEAN ABSOLUTE PERCENTAGE ERROR")
# No Scaling
print("MAPE                :", mape(y_test, pred))

# Standard Scaling
print("Standard Scaling")
print("MAPE (On Scale)     :", mape(y_test_z, pred_z))
print("MAPE (Inverse Scale):", mape(z_scale_y.inverse_transform(y_test_z), pred_z_inverse))

# Minmax Scaling
print("Minmax Scaling")
print("MAPE (On Scale)     :", mape(y_test_minmax, pred_minmax))
print("MAPE (Inverse Scale):", mape(minmax_scale_y.inverse_transform(y_test_minmax), pred_minmax_inverse))

# R2
from sklearn.metrics import r2_score as r2
print("R-Squared")
# No Scaling
print("R2                :", r2(y_test, pred))

# Standard Scaling
print("Standard Scaling")
print("R2 (On Scale)     :", r2(y_test_z, pred_z))
print("R2 (Inverse Scale):", r2(z_scale_y.inverse_transform(y_test_z), pred_z_inverse))

# Minmax Scaling
print("Minmax Scaling")
print("R2 (On Scale)     :", r2(y_test_minmax, pred_minmax))
print("R2 (Inverse Scale):", r2(minmax_scale_y.inverse_transform(y_test_minmax), pred_minmax_inverse))
```
